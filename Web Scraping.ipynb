```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://wuzzuf.net/search/jobs/?q=data+scinese&a=hpb"
```

```python
client = urlopen(url)
```

```python
html = client.read()
html
```

```python
client.close()
```

```python
soup = BeautifulSoup(html , "html.parser")
soup
```

```python
containers = soup.find_all("div" , {"class":"css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
BeautifulSoup.prettify(containers[0])
```

```python
f = open("wuzzef.csv", "w")
f.write("Jop_title, company, Jop_type\n")
```

```python
for container in containers:
    
    Jop_title = container.findAll("a", {"class": "css-o171kl"})[0].text.strip()
    
    company = container.findAll("a", {"class": "css-17s97q8"})[0].text.strip()
    
    addrs = container.findAll("span", {"class": "css-5wys0k"})[0].text.strip()
    
    Jop_type = container.findAll("span", {"class": "css-1ve4b75 eoyjyou0"})[0].text.strip()
    
    
    
#    print(Jop_title)
#    print(company)
#    print(addrs)
#    print(Jop_type)
#    print(Jop_title + ", " + company + ", " + addrs + ", "+ Jop_type)
#    print()
    
    
    f.write(Jop_title + ", " + company +", "+ Jop_type + "\n")
f.close()
```
